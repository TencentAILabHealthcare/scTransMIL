```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./data/time_res.csv")
df['log_time'] = np.log10(df['time(s)'] + 1)

a = 1.5
b = 2
title = "Inference time"

###################################################################
sns.set_theme(style="ticks", 
              font_scale=1.2
              )

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(a, b))
#ax.set_xscale("log")
plt.rc('font', size=16)

# Plot the orbital period with horizontal boxes
sns.barplot(
    df, y="log_time", x="Method", hue="Method", 
    # palette="light:b"
    # whis=[0, 100], width=.6
)

ax.set_xlabel("", fontdict={"fontsize":16})
ax.set_ylabel("log(time(s)+1)", fontdict={"fontsize":16})
ax.set_title(title, fontdict={"fontsize":16})

for tick in ax.get_xticklabels():
   tick.set_rotation(90)
sns.despine()
```
