```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("./data/multicls-pancancer.csv")

a = 1.5
b = 2
title = "Internal dataset"

###################################################################
sns.set_theme(style="ticks", 
              font_scale=1.2
              )

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(a, b))
#ax.set_xscale("log")
plt.rc('font', size=16)

# Plot the orbital period with horizontal boxes
sns.boxplot(
    df, y="acc", x="Method", hue="Method", 
    # palette="light:b"
    # whis=[0, 100], width=.6
)

# Add in points to show each observation
# sns.stripplot(df, y="acc", x="Method", hue="Method", size=4, color=".3")

# Tweak the visual presentation
#ax.xaxis.grid(True)
ax.set_xlabel("", fontdict={"fontsize":16})
ax.set_ylabel("Acc", fontdict={"fontsize":16})
ax.set_title(title, fontdict={"fontsize":16})

plt.ylim(0.4, 0.9)

for tick in ax.get_xticklabels():
   tick.set_rotation(90)
sns.despine()
```
